# load and initialize the dataset

```python
import os
from glob import glob
from PIL import Image

data_dir = "../data/images"  # This folder is gitignored; local only
classes = sorted(os.listdir(data_dir))

print("Found classes:", classes)
for cls in classes:
    class_path = os.path.join(data_dir, cls)
    images = glob(os.path.join(class_path, "*.jpg"))
    print(f"Class '{cls}' has {len(images)} images.")

# Optional: show a sample image
import matplotlib.pyplot as plt

sample_class = classes[0]
sample_image_path = glob(os.path.join(data_dir, sample_class, "*.jpg"))[0]
img = Image.open(sample_image_path)
plt.imshow(img)
plt.title(f"Sample from class: {sample_class}")
plt.show()
```
